```python
#exctracting the data from the given URl and stored in all_records[].
all_records=[]
import requests
import time
for page in range(1,6):
  url=f"https://api.coingecko.com/api/v3/coins/markets?vs_currency=inr&per_page=250&order=market_cap_desc&page={page}&sparkline=False"
  response = requests.get(url)
  if response.status_code == 200:
    data=response.json()
    all_records.extend(data)
```

```python
len(all_records)
```

```python
#renamning the column name and assigning them.
Records=[]
for i in all_records:
  Records.append(dict(Id=i['id'],Symbol=i['symbol'],Name=i['name'],Current_price=i['current_price'],Market_cap=i['market_cap'],Rank=i['market_cap_rank'],Total_volume=i['total_volume'],Circulating_supply=i['circulating_supply'],Total_supply=i['total_supply'],Ath=i['ath'],Atl=i['atl'],Last_updated=i['last_updated']))
Records
```

```python
#importing pandas and view all the records for clear understanding and for proper allignment.
import pandas as pd
coins_df=pd.DataFrame(Records)
coins_df
```

```python
#viewing only the Last_updated column
coins_df['Last_updated'] = pd.to_datetime(coins_df['Last_updated'])
coins_df['Last_updated']
```

```python
#exacting only date from the timestamp as required
coins_df['Date'] = coins_df['Last_updated'].dt.date
coins_df['Date']
```

```python
#view the coins dataframe and can find the extended cloumn with name date_only
coins_df
```

```python
#dropping the Last_updated column as it is not required
coins_df = coins_df.drop('Last_updated', axis=1, errors='ignore')
coins_df
```

```python
#scrolling through the data from the URL And exracting the coins history for last 365 days from the day of running the code.
coins_data=[]
coins_ids=['bitcoin','ethereum','tether','binancecoin','ripple']
for i in coins_ids:
  url=f"https://api.coingecko.com/api/v3/coins/{i}/market_chart?vs_currency=inr&days=365"
  response=requests.get(url)
  if response.status_code == 200:
    time.sleep(5)
    coins=response.json()
    prices = coins["prices"]#exacting data to table and converting ms to datestamp
    for ts, price in prices:
      coins_data.append({"coin_id": i,"date": pd.to_datetime(ts, unit="ms"),"price": price})
coinsPrice_df = pd.DataFrame(coins_data)

coinsPrice_df
```

```python
#exacting only date from the time stamp of coinsPrice_df dataframe
coinsPrice_df['Date'] = coinsPrice_df['date'].dt.date
coinsPrice_df['Date']
```

```python
#dropping the old date column

coinsPrice_df =coinsPrice_df.drop('date', axis=1, errors='ignore')
coinsPrice_df
```

#Creating Tables


```python
#creating table for cryptocurrencies:
import sqlite3

conn = sqlite3.connect("crtpo.db")
cursor = conn.cursor()

conn.execute("""CREATE TABLE Cryptocurrencies
                 (id VARCHAR(50) PRIMARY KEY,
                 symbol VARCHAR(10),
                 name VARCHAR(100),
                 current_price DECIMAL(18, 6),
                 market_cap BIGINT,
                 market_cap_rank INT,
                 total_volume BIGINT,
                 circulating_supply DECIMAL(20, 6),
                 total_supply DECIMAL(20, 6),
                 ath DECIMAL(18, 6),
                 atl DECIMAL(18, 6),
                 date DATE
                 );

                 """)
conn.commit()
conn.close()
```

```python
#creating table for Crypto_prices
import sqlite3
conn = sqlite3.connect("crtpo.db")
cursor = conn.cursor()

query=conn.execute("""CREATE TABLE Crypto_prices
                 (coin_id VARCHAR(50),
                 date DATE,
                 price DECIMAL(18, 6)
                 );

                 """)
conn.commit()
print(query)
```

```python
#creating table for Oil_prices

import sqlite3
conn = sqlite3.connect("crtpo.db")
cursor = conn.cursor()

query=conn.execute("""CREATE TABLE Oil_prices
                 (date DATE PRIMARY KEY,
                 price_INR DECIMAL(18, 6)
                 );

                 """)
conn.commit()
print(query)
```

```python
#creating table for Stock_prices

import sqlite3
conn = sqlite3.connect("crtpo.db")
cursor = conn.cursor()

query=conn.execute("""CREATE TABLE Stock_prices
                 (date DATE,
                 open DECIMAL(18, 6),
                 high DECIMAL(18, 6),
                 low DECIMAL(18, 6),
                 close DECIMAL(18, 6),
                 volume BIGINT,
                 ticker VARCHAR(20)
                 );

                 """)
conn.commit()
print(query)
```

#oil_prices

```python
#1. Find the highest oil price in the last 5 years.
df = pd.read_sql("""SELECT MAX(Price) AS highest_price
FROM Oil_prices
WHERE Date >= (
  SELECT substr(MAX(date), 1, 4) - 5 || '-01-01'
  FROM Oil_prices
);

""", conn)
df
```

```python
#2. Get the average oil price per year.
df = pd.read_sql("""SELECT
  substr(date, 1, 4) AS year,
  AVG(Price) AS avg_price
FROM Oil_prices
GROUP BY substr(date, 1, 4)
ORDER BY year;

""", conn)
df
```

```python
#3. Show oil prices during COVID crash (MarchApril 2020).
df = pd.read_sql("""
SELECT *FROM Oil_prices
WHERE date BETWEEN '2020-03-01' AND '2020-04-30'
ORDER BY date;

""", conn)
df
```

```python
#4. Find the lowest price of oil in the last 10 years.
df = pd.read_sql("""SELECT MIN(Price) AS lowest_price
FROM Oil_prices
WHERE date >= (
  SELECT substr(MAX(date), 1, 4) - 10 || '-01-01'
  FROM Oil_prices
);
""", conn)
df
```

```python
#5. Calculate the volatility of oil prices (max-min difference per year).
df = pd.read_sql("""SELECT
  substr(date, 1, 4) AS year,
  MAX(Price) - MIN(Price) AS volatility
FROM Oil_prices
GROUP BY substr(date, 1, 4)
ORDER BY year;

""", conn)
df
```

```python
#1. Compare Bitcoin vs Oil average price in 2025.

df = pd.read_sql("""SELECT
  'bitcoin' AS asset,
  AVG(price) AS avg_price
FROM crypto_prices
WHERE coin_id = 'bitcoin'
  AND strftime('%Y', date) = '2025'

UNION ALL

SELECT
  'OIL' AS asset,
  AVG(price) AS avg_price
FROM oil_prices
WHERE strftime('%Y', date) = '2025';

""", conn)
df
```

```python
#2. Check if Bitcoin moves with S&P 500 (correlation idea).

df=pd.read_sql("""SELECT
    date(c.date) AS date,
    c.price AS btc_close,
    s.close AS sp500_close
FROM crypto_prices c
LEFT JOIN stock_prices s
    ON date(c.date) = date(s.date)
   AND s.ticker = '^GSPC'
WHERE c.coin_id = 'bitcoin'
ORDER BY date;
""",conn)
df
```

```python
#3. Compare Ethereum and NASDAQ daily prices for 2025.
df = pd.read_sql("""
SELECT
    date(c.date) AS date,
    c.price AS eth_price,
    s.close AS nasdaq_close
FROM crypto_prices c
JOIN stock_prices s
    ON date(c.date) = date(s.date)
WHERE c.coin_id = 'ethereum'
  AND s.ticker = '^IXIC'
  AND strftime('%Y', c.date) = '2025'
ORDER BY date;
""", conn)

df
```

```python
#4. Find days when oil price spiked and compare with Bitcoin price change.

df = pd.read_sql("""
SELECT
    date(o.Date) AS date,
    o.Price AS oil_price,
    b.price AS btc_price
FROM oil_prices o
JOIN crypto_prices b
    ON date(o.Date) = date(b.Date)
WHERE b.coin_id = 'bitcoin'
ORDER BY date;
""", conn)
df
```

```python
#5. Compare top 3 coins daily price trend vs Nifty (^NSEI).
df = pd.read_sql("""
SELECT
    date(c.Date) AS date,
    c.coin_id AS crypto,
    c.price AS crypto_price,
    n.Close AS nifty_close
FROM crypto_prices c
JOIN stock_prices n
    ON date(c.Date) = date(n.Date)
WHERE c.coin_id IN ('bitcoin', 'ethereum', 'binancecoin')
  AND n.Ticker = '^NSEI'
ORDER BY date, crypto;
""", conn)
df
```

```python
#6. Compare stock prices (^GSPC) with crude oil prices on the same dates


df = pd.read_sql("""
SELECT
    date(s.Date) AS date,
    s.Close AS sp500_close,
    o.Price AS oil_price
FROM stock_prices s
JOIN oil_prices o
    ON date(s.Date) = date(o.Date)
WHERE s.Ticker = '^GSPC'
ORDER BY date;
""", conn)

df
```

```python
#7. Correlate Bitcoin closing price with crude oil closing price (same date)

df = pd.read_sql("""
SELECT
    date(b.Date) AS date,
    b.price AS btc_price,
    o.Price AS oil_price
FROM crypto_prices b
JOIN oil_prices o
    ON date(b.Date) = date(o.Date)
WHERE b.coin_id = 'bitcoin'
ORDER BY date;
""", conn)

df
```

```python
#8. Compare NASDAQ (^IXIC) with Ethereum price trends

df = pd.read_sql("""
SELECT
    date(e.Date) AS date,
    e.price AS eth_price,
    n.Close AS nasdaq_close
FROM crypto_prices e
JOIN stock_prices n
    ON date(e.Date) = date(n.Date)
WHERE e.coin_id = 'ethereum'
  AND n.Ticker = '^IXIC'
ORDER BY date;
""", conn)

df
```

```python
#9. Join top 3 crypto coins with stock indices for 2025
df = pd.read_sql("""
SELECT
    date(c.Date) AS date,
    c.coin_id AS crypto,
    c.price AS crypto_price,
    s.Ticker AS index_symbol,
    s.Close AS index_close
FROM crypto_prices c
JOIN stock_prices s
    ON date(c.Date) = date(s.Date)
WHERE c.coin_id IN ('bitcoin', 'ethereum', 'binancecoin')
  AND s.Ticker IN ('^GSPC', '^IXIC', '^NSEI')
  AND strftime('%Y', c.Date) = '2025'
ORDER BY date, crypto;
""", conn)

df
```

```python
#10. Multi-join: stock prices, oil prices, and Bitcoin prices for daily comparison

df = pd.read_sql("""
SELECT
    date(b.Date) AS date,
    b.price AS btc_price,
    s.Close AS sp500_close,
    o.Price AS oil_price
FROM crypto_prices b
JOIN stock_prices s
    ON date(b.Date) = date(s.Date)
JOIN oil_prices o
    ON date(b.Date) = date(o.Date)
WHERE b.coin_id = 'bitcoin'
  AND s.Ticker = '^GSPC'
ORDER BY date;
""", conn)

df
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import sqlite3
import plotly.express as px
from datetime import datetime

st.set_page_config(page_title="Cross-Market Analysis", layout="wide")
st.title(" Cross-Market Analysis: Crypto, Oil & Stocks")
# LOAD CSVs
@st.cache_data
def load_data():
    coins_df = pd.read_csv("coins.csv")
    coinsPrice_df = pd.read_csv("coinsPrice.csv")
    oil_df = pd.read_csv("oil.csv")
    stocks_df = pd.read_csv("AllStocks.csv")

    coinsPrice_df['Date'] = pd.to_datetime(coinsPrice_df['Date'])
    oil_df['Date'] = pd.to_datetime(oil_df['Date'])
    stocks_df['Date'] = pd.to_datetime(stocks_df['Date'])

    return coins_df, coinsPrice_df, oil_df, stocks_df

coins_df, coinsPrice_df, oil_df, stocks_df = load_data()

#CREATE SQLITE DB
@st.cache_resource
def create_db():
    conn = sqlite3.connect("crypto.db")
    coins_df.to_sql("cryptocurrencies", conn, if_exists="replace", index=False)
    coinsPrice_df.to_sql("crypto_prices", conn, if_exists="replace", index=False)
    oil_df.to_sql("oil_prices", conn, if_exists="replace", index=False)
    stocks_df.to_sql("stock_prices", conn, if_exists="replace", index=False)
    return conn

conn = create_db()

# SIDEBAR
page = st.sidebar.selectbox(
    "Select Page",
    ["Filters & Data Exploration", "SQL Query Runner", "Top Crypto Analysis"]
)

# PAGE 1: Filters & Data Exploration
if page == "Filters & Data Exploration":
    st.subheader(" Market Overview")
    start_date = st.date_input("Start Date", datetime(2025, 1, 1))
    end_date = st.date_input("End Date", datetime(2026, 1, 1))

    # convert to datetime64 for comparison
    start_dt = pd.to_datetime(start_date)
    end_dt = pd.to_datetime(end_date)

    btc_df = coinsPrice_df[
        (coinsPrice_df["coin_id"] == "bitcoin") &
        (coinsPrice_df["Date"].between(start_dt, end_dt))
    ]

    oil_filtered = oil_df[oil_df["Date"].between(start_dt, end_dt)]
    sp500 = stocks_df[(stocks_df["Ticker"] == "^GSPC") & (stocks_df["Date"].between(start_dt, end_dt))]
    nifty = stocks_df[(stocks_df["Ticker"] == "^NSEI") & (stocks_df["Date"].between(start_dt, end_dt))]

    col1, col2, col3, col4 = st.columns(4)
    col1.metric("Bitcoin Avg Price", round(btc_df["price"].mean(), 2))
    col2.metric("Oil Avg Price", round(oil_filtered["Price"].mean(), 2))
    col3.metric("S&P 500 Avg Close", round(sp500["Close"].mean(), 2))
    col4.metric("NIFTY Avg Close", round(nifty["Close"].mean(), 2))

    snapshot = (
        btc_df.merge(oil_filtered, on="Date", how="inner", suffixes=("_btc","_oil"))
        .merge(sp500[["Date","Close"]], on="Date")
        .merge(nifty[["Date","Close"]], on="Date", suffixes=("_sp500","_nifty"))
    )

    st.subheader(" Daily Market Snapshot")
    st.dataframe(snapshot[["Date","price","Price","Close_sp500","Close_nifty"]])

#  PAGE 2: SQL Query Runner
elif page == "SQL Query Runner":
    st.subheader(" SQL Query Runner")

    # Queries Categorization
    queries = {
        "cryptocurrencies": {
            "Top 3 by Market Cap": "SELECT * FROM cryptocurrencies ORDER BY Market_cap DESC LIMIT 3;",
            "Circulating supply > 90% of total supply": "SELECT * FROM cryptocurrencies WHERE Circulating_supply > 0.9 * Total_supply;",
            "Within 10% of ATH": "SELECT * FROM cryptocurrencies WHERE Current_price >= 0.9 * Ath;",
            "Average Rank for volume > $1B": "SELECT AVG(Rank) AS avg_Rank FROM cryptocurrencies WHERE Total_volume > 1000000000;",
            "Most recently updated coin": "SELECT * FROM cryptocurrencies ORDER BY Date DESC LIMIT 1;"
        },
        "crypto_prices": {
            "Highest daily BTC price last 365 days": "SELECT * FROM crypto_prices WHERE coin_id='bitcoin' AND Date >= date('now','-365 days') ORDER BY Date DESC LIMIT 1;",
            "Average ETH daily price past year": "SELECT AVG(price) AS avg_daily_price FROM crypto_prices WHERE coin_id='ethereum' AND Date >= date('now','-365 days');",
            "BTC daily trend Feb 2025": "SELECT * FROM crypto_prices WHERE coin_id='bitcoin' AND Date BETWEEN '2025-02-01' AND '2025-02-28' ORDER BY Date;",
            "Highest avg coin price 1Y": "SELECT coin_id, AVG(price) AS avg_price_1y FROM crypto_prices WHERE Date >= date('now','-1 year') GROUP BY coin_id ORDER BY avg_price_1y DESC LIMIT 1;",
            "BTC % change Feb 2025Feb 2026": """SELECT ((p2026.price - p2025.price)*100.0/p2025.price) AS pct_change
                                               FROM crypto_prices p2025
                                               JOIN crypto_prices p2026 ON p2025.coin_id=p2026.coin_id
                                               WHERE p2025.coin_id='bitcoin' AND p2025.date='2025-02-05' AND p2026.date='2026-02-04';"""
        },
        "oil_prices": {
            "Highest oil price last 5 years": "SELECT MAX(Price) AS highest_price FROM oil_prices WHERE Date >= (SELECT substr(MAX(Date),1,4)-5||'-01-01' FROM oil_prices);",
            "Average oil price per year": "SELECT substr(Date,1,4) AS year, AVG(Price) AS avg_price FROM oil_prices GROUP BY substr(Date,1,4) ORDER BY year;",
            "Oil prices COVID crash Mar-Apr 2020": "SELECT * FROM oil_prices WHERE Date BETWEEN '2020-03-01' AND '2020-04-30' ORDER BY Date;",
            "Lowest oil price last 10 years": "SELECT MIN(Price) AS lowest_price FROM oil_prices WHERE Date >= (SELECT substr(MAX(Date),1,4)-10||'-01-01' FROM oil_prices);",
            "Oil volatility per year": "SELECT substr(Date,1,4) AS year, MAX(Price)-MIN(Price) AS volatility FROM oil_prices GROUP BY substr(Date,1,4) ORDER BY year;"
        },
        "stock_prices": {
            "All prices for ^IXIC/^NSEI/^GSPC": "SELECT * FROM stock_prices WHERE Ticker IN ('^IXIC','^NSEI','^GSPC') ORDER BY Date;",
            "Highest NASDAQ close": "SELECT MAX(Close) AS highest_close FROM stock_prices WHERE Ticker='^IXIC';",
            "Top 5 high-low diff S&P 500": "SELECT Date,(High-Low) AS price_difference FROM stock_prices WHERE Ticker='^GSPC' ORDER BY price_difference DESC LIMIT 5;",
            "Monthly avg close per ticker": "SELECT Ticker, substr(Date,1,7) AS month, AVG(Close) AS avg_monthly_close FROM stock_prices GROUP BY Ticker, substr(Date,1,7) ORDER BY Ticker, month;",
            "NSEI avg volume 2024": "SELECT AVG(Volume) AS avg_volume_2024 FROM stock_prices WHERE Ticker='^NSEI' AND Date BETWEEN '2024-01-01' AND '2024-12-31';"
        },
        "join_queries": {
            "Bitcoin vs Oil avg 2025": """SELECT 'bitcoin' AS asset, AVG(price) AS avg_price FROM crypto_prices WHERE coin_id='bitcoin' AND strftime('%Y',Date)='2025'
                                          UNION ALL SELECT 'OIL' AS asset, AVG(Price) AS avg_price FROM oil_prices WHERE strftime('%Y',Date)='2025';""",
            "BTC correlation with S&P500": """SELECT date(c.Date) AS date, c.price AS btc_close, s.Close AS sp500_close FROM crypto_prices c
                                              LEFT JOIN stock_prices s ON date(c.Date)=date(s.Date) AND s.Ticker='^GSPC'
                                              WHERE c.coin_id='bitcoin' ORDER BY date;""",
            "ETH vs NASDAQ daily 2025": """SELECT date(c.Date) AS date, c.price AS eth_price, s.Close AS nasdaq_close FROM crypto_prices c
                                           JOIN stock_prices s ON date(c.Date)=date(s.Date)
                                           WHERE c.coin_id='ethereum' AND s.Ticker='^IXIC' AND strftime('%Y',c.Date)='2025'
                                           ORDER BY date;""",
            "Oil spike vs BTC price": """SELECT date(o.Date) AS date, o.Price AS oil_price, b.price AS btc_price FROM oil_prices o
                                         JOIN crypto_prices b ON date(o.Date)=date(b.Date)
                                         WHERE b.coin_id='bitcoin' ORDER BY date;""",
            "Top 3 coins vs Nifty": """SELECT date(c.Date) AS date, c.coin_id AS crypto, c.price AS crypto_price, n.Close AS nifty_close
                                        FROM crypto_prices c JOIN stock_prices n ON date(c.Date)=date(n.Date)
                                        WHERE c.coin_id IN ('bitcoin','ethereum','binancecoin') AND n.Ticker='^NSEI' ORDER BY date, crypto;""",
            "S&P500 vs Oil": """SELECT date(s.Date) AS date, s.Close AS sp500_close, o.Price AS oil_price FROM stock_prices s
                               JOIN oil_prices o ON date(s.Date)=date(o.Date) WHERE s.Ticker='^GSPC' ORDER BY date;""",
            "BTC vs Oil closing price": """SELECT date(b.Date) AS date, b.price AS btc_price, o.Price AS oil_price FROM crypto_prices b
                                          JOIN oil_prices o ON date(b.Date)=date(o.Date) WHERE b.coin_id='bitcoin' ORDER BY date;""",
            "NASDAQ vs ETH": """SELECT date(e.Date) AS date, e.price AS eth_price, n.Close AS nasdaq_close FROM crypto_prices e
                               JOIN stock_prices n ON date(e.Date)=date(n.Date) WHERE e.coin_id='ethereum' AND n.Ticker='^IXIC' ORDER BY date;""",
            "Top 3 coins vs indices 2025": """SELECT date(c.Date) AS date, c.coin_id AS crypto, c.price AS crypto_price, s.Ticker AS index_symbol, s.Close AS index_close
                                              FROM crypto_prices c JOIN stock_prices s ON date(c.Date)=date(s.Date)
                                              WHERE c.coin_id IN ('bitcoin','ethereum','binancecoin') AND s.Ticker IN ('^GSPC','^IXIC','^NSEI') AND strftime('%Y',c.Date)='2025'
                                              ORDER BY date, crypto;""",
            "Multi-join BTC+S&P+Oil": """SELECT date(b.Date) AS date, b.price AS btc_price, s.Close AS sp500_close, o.Price AS oil_price
                                         FROM crypto_prices b JOIN stock_prices s ON date(b.Date)=date(s.Date)
                                         JOIN oil_prices o ON date(b.Date)=date(o.Date)
                                         WHERE b.coin_id='bitcoin' AND s.Ticker='^GSPC' ORDER BY date;"""
        }
    }

    category = st.selectbox("Select Query Category", list(queries.keys()))
    query_name = st.selectbox("Select Query", list(queries[category].keys()))

    if st.button("Run Query"):
        sql = queries[category][query_name]
        df = pd.read_sql(sql, conn)
        st.dataframe(df)

        # For join_queries, show line chart automatically
        if category=="join_queries":
            numeric_cols = df.select_dtypes(include='number').columns
            if len(numeric_cols) > 1:
                fig = px.line(df, x=df.columns[0], y=numeric_cols, title=query_name)
                st.plotly_chart(fig, use_container_width=True)

# PAGE 3: Top Crypto Analysis
else:
    st.subheader(" Top Crypto Price Trend")
    top_coins = coins_df.sort_values("Rank").head(3)["Id"].tolist()
    selected_coin = st.selectbox("Select Coin", top_coins)

    start = st.date_input("From", datetime(2024, 1, 1), key="c1")
    end = st.date_input("To", datetime(2025, 1, 1), key="c2")

    start_dt = pd.to_datetime(start)
    end_dt = pd.to_datetime(end)

    selected_coin_df = coinsPrice_df[
        (coinsPrice_df["coin_id"]==selected_coin) &
        (coinsPrice_df["Date"].between(start_dt, end_dt))
    ]

    fig = px.line(selected_coin_df, x="Date", y="price", title=f"{selected_coin.upper()} Price Trend")
    st.plotly_chart(fig, use_container_width=True)
    st.dataframe(selected_coin_df)
```
